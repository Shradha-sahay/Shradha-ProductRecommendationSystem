```python
tfidf = TfidfVectorizer(stop_words='english')
des_tfidf = tfidf.fit_transform(df['product_name'])
des_tfidf.shape
```

```python
des_cosine_sim=cosine_similarity(des_tfidf,des_tfidf)
des_cosine_sim.shape
```

```python
def recommend(category, sim):
    products = df['category'].values

    product_idx = np.where(products == category)[0]
    similarity_scores = sim[product_idx].ravel()
    products_and_scores = dict(zip(products, similarity_scores)).items()
    top_10 = sorted(products_and_scores, key= lambda x: x[1], reverse= True)[1:11]

    idx = [product_idx[0]]
    for i, (name, score) in enumerate(top_10):
        idx.append(list(np.where(products == name)[0])[0])
    return df.iloc[idx]
```

```python
recommend('Computers&Accessories|Accessories&Peripherals|Cables&Accessories|Cables|USBCables', des_cosine_sim)
```
